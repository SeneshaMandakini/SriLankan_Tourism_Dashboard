#Load and Split data

```python
import pandas as pd
```

```python
#Import numpy for performing the statistical and mathematical operations on the data set
import numpy as np
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
TA = pd.read_excel('combined_df.xlsx')
```

```python
TA
```

```python
np.random.seed(0)
```

```python
# Drop rows with NaN values
TA = TA.dropna()
```

```python
TA
```

```python
TA['Tourist arrivals'] = np.log(TA['Tourist arrivals'])
```

```python
tscv = TimeSeriesSplit(n_splits=5)
```

```python
train_percentage = 0.8
```

```python
split_index = int(len(TA) * train_percentage)
```

```python
train_data = TA.iloc[:split_index]
test_data = TA.iloc[split_index:]
```

```python
test_data.head()
```

```python
Train_year = train_data[['Year','Month']]
Test_year = test_data[['Year','Month']]
```

```python
Test_year
```

#Past Tourism Arrivals

```python
X_train_1, y_train_1= train_data[['lag_1','lag_2', 'lag_3']], train_data['Tourist arrivals']
X_test_1, y_test_1 = test_data[['lag_1','lag_2', 'lag_3']], test_data['Tourist arrivals']
```

```python
# Initialize MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Fit the scaler on the training data and transform it
X_train_scaled_1 = scaler.fit_transform(X_train_1)

# Transform the testing data
X_test_scaled_1 = scaler.transform(X_test_1)
```

```python
rf_model_past = RandomForestRegressor(random_state=42)
rf_model_past.fit(X_train_scaled_1, y_train_1)
```

```python
# rf_model is your trained random forest model
params_past = rf_model_past.get_params()
```

```python
# Print the parameters
print("Parameters used in the Random Forest model:")
print(params_past)
```

```python
rf_model_past.score(X_train_scaled_1, y_train_1)
```

```python
y_pred_1 = rf_model_past.predict(X_test_scaled_1)
```

```python
np.random.seed(0)
```

```python
y_pred_1
```

```python
#Mean Squared Error (MSE)
mse_1= mean_squared_error(y_test_1, y_pred_1)
print(f'Mean Squared Error (MSE): {mse_1}')
```

```python
y_test_antilog = np.exp(y_test_1)
y_pred_antilog = np.exp(y_pred_1)
```

```python
# Create a DataFrame with the test data, actual values, and predicted values
test_results_1 = pd.DataFrame(X_test_scaled_1, columns=[f'feature_{i}' for i in range(X_test_scaled_1.shape[1])])
test_results_1['Actual'] = y_test_antilog
test_results_1['Predicted'] = y_pred_antilog
```

```python
y_test_antilog
```

```python
y_pred_antilog
```

```python
y_pred_antilog
```

```python
y_pred_1
```

```python
test_results_1
```

```python
test_results_1.to_csv('Tourism_Past.csv', index=False)
```

##RF after tuning parameters

```python
# Set random seed for NumPy
np.random.seed(0)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Define hyperparameters to search
param_grid_rf_1 = {
    'n_estimators': [50, 100, 150, 200],
    'max_depth': [3, 5, 7, 10, None],
    'min_samples_split': [2, 4, 6, 8, 10],
    'min_samples_leaf': [1, 2, 3, 4],
    'max_features': ['sqrt', 'log2', None, 2, 4, 6, 8, 10],
    'bootstrap': [True, False]
}
```

```python
# Perform Grid Search with 5-fold cross-validation
grid_search_rf_1 = GridSearchCV(estimator=rf_model_past, param_grid=param_grid_rf_1, cv=tscv, scoring='neg_mean_squared_error')
grid_search_rf_1.fit(X_train_scaled_1, y_train_1)
```

```python
# Get the best parameters
best_params_rf_1 = grid_search_rf_1.best_params_
```

```python
# Train the model with the best parameters
best_rf_model_1 = RandomForestRegressor(**best_params_rf_1, random_state=42)
best_rf_model_1.fit(X_train_scaled_1, y_train_1)
```

```python
best_rf_model_1.score(X_train_scaled_1, y_train_1)
```

```python
# Make predictions and evaluate the model
y_pred_best_rf_1 = best_rf_model_1.predict(X_test_scaled_1)
mse_best_rf_1 = mean_squared_error(y_test_1, y_pred_best_rf_1)
```

```python
print(f'Best Parameters: {best_params_rf_1}')
print(f'Mean Squared Error (MSE) with Best Model: {mse_best_rf_1}')
```

```python
y_test_antilog = np.exp(y_test_1)
y_pred_antilog = np.exp(y_pred_best_rf_1)
```

```python
# Create a DataFrame with the test data, actual values, and predicted values
test_results_1 = pd.DataFrame(X_test_scaled_1, columns=[f'feature_{i}' for i in range(X_test_scaled_1.shape[1])])
test_results_1['Actual'] = y_test_antilog
test_results_1['Predicted'] = y_pred_antilog
```

```python
y_pred_antilog
```

```python
y_pred_1
```
